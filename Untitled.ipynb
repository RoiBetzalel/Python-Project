### The Speedruns Route
A typical Minecraft speedrun starts with spawning in and gathering resources such as food, iron (for a bucket and pickaxe), flint and steel, and a lava source for obsidian to enter the Nether.

In the Nether, you are looking for two things: a Bastion and a Fortress.

In the Bastion, you trade gold for items such as fire resistance potions, obsidian, ender pearls, and string to craft beds to defeat the Ender Dragon.

In the Fortress, you kill Blazes for blaze rods, which are used to craft Eyes of Ender to locate the Stronghold and reach the End.

After obtaining the Eyes of Ender, you create a Nether portal using obsidian from the Bastion and calculate how far the Stronghold is. If it is close, you travel there directly. If it is far, you return to the Nether and build another portal closer to the target location.

After reaching the Stronghold, you search for the portal room and fill the End Portal with Eyes of Ender. This transports you to the End, where you fight the Ender Dragon. After defeating her, you pass through the End Portal and finish the run.

### Starting the Game:
When starting a world, speedrunners try to find the fastest way to obtain an iron source and enter the Nether.
#### Iron Sources:
There are several ways to obtain iron quickly:

“Structureless” – The most common method. You typically look for a cave with enough iron.

“Shipwreck” – If you spawn near an ocean, you’ll try to spot a shipwreck. A full shipwreck can contain a food chest, a map leading to buried treasure (“BT”), and an iron chest.
“Half shipwreck” – Similar, but without the food chest.

“Village” – In almost every biome, there is a village you can go to. When entering a village, you try to either kill an Iron Golem (which can drop 3–5 iron), trade with a villager for an iron bucket, or get lucky and find a blacksmith with enough iron and other resources. Villages usually have a lot of hay bales, which can be turned into bread.

“Buried Treasure” – If you spawn on an island, you can use a trick called “mapless” to find it quickly. However, the odds of finding one are very low, and even if you do, it’s not guaranteed you’ll get enough resources to continue the run. (The TNT can be used as an easier way to gather wood and blocks for traversing the Nether.)

“Desert Temple” – There are four chests at the bottom that can contain iron and food. The temple also has a trap with nine TNT blocks that can be collected and used to gather blocks more easily.

“Nether” – Obtaining iron from Bastion chests or from a ruined portal in the Nether.

“Misc” – This usually includes a ruined portal (which may have a chest with iron), a jungle temple, or a pillager outpost.

# Bastion Types
#### There are four types of Bastions:
Housing, Bridge, Treasure, and Stables.  

Before speedrunners learned how to properly use Bastions, they used to mine gold nuggets and craft gold ingots to trade with Piglins. This method of trading is called “Classic”.  

When Couriway speedruns on newer versions of the game, the chance of obtaining Ender Pearls from Piglin bartering and Bastion loot is much lower due to updates to Piglin trade rates. Because of this, he tends to hunt Endermen instead, since they have about a 50% chance of dropping an Ender Pearl.

```python
#making dummies for the bastions

# cleating text
df["Bastion Type"] = df["Bastion Type"].str.lower().str.strip()
df["Bastion Type"] = df["Bastion Type"].str.replace(" / ", "/")

# mapping categories
mapping = {
    # bastions
    "treasure": "Treasure",
    "stables": "Stables",
    "bridge": "Bridge",
    "housing": "Housing",

    # no trade / no bastion
    "warped forest": "Hunting Enderman",
    
    # classic routes
    "classic": "Classic",
}

# creating dummies
dummies = df["Bastion Type"].str.get_dummies(sep="/")

# mapping
dummies = dummies.rename(columns=mapping)

# תיקון ל-Pandas חדש
dummies = dummies.T.groupby(level=0).max().T

# חיבור
df = pd.concat([df, dummies], axis=1)
```

# The Ends Dimention
The End dimension is the third and final dimension in the game, where there are 10 pillars of different heights (random locations each seed, but not random heights). The dimension is filled with Endermen, and each pillar has a crystal on top that heals the final boss, the Ender Dragon.

The way the developers intended players to beat the game was by using bows, arrows, and swords to defeat the final boss. However, speedrunners found several ways to defeat her faster.
### End Fight strats
#### easiest to hardest
Other – This usually means the player ran out of beds, and the fight took several perches.

Perch – A strategy where the player waits for the Ender Dragon to perch toward the fountain in the middle of the dimension, then uses beds to explode and defeat her.

Half Bow – A strategy where the player destroys 8 of the crystals and then waits for the dragon to perch toward the fountain to use beds to deal damage.

Full Bow – The same as Half Bow, but the player destroys all 10 crystals.

Ground Zero – A strategy where the player manipulates the dragon to circle around them without hitting the player, allowing them to use beds safely to damage her.

Zero Cycle – A newer strategy where the player builds up to one of the pillars and uses beds to defeat the dragon quickly.

```python
# dictionary
fix_map = {
    "Half bow": "Half Bow",
    "Full bow": "Full Bow",
    "Ful Bow": "Full Bow",
    "other": "Other",
    "Ground Zero ": "Ground Zero"
}

# swapping
df['End Fight Type'] = df['End Fight Type'].replace(fix_map)

# clearing
df['End Fight Type'] = df['End Fight Type'].str.strip().str.title()

print(df['End Fight Type'].value_counts())
```

```python
#new column for only the end fight time
df["End Fight Time"] = df["IGT"] - df["End"]
#turning it to minutes
df["End Fight Minutes"] = df["End Fight Time"].dt.total_seconds() / 60
```

Sometimes you don’t get enough obsidian to build a second portal, or you get lucky and end up very close to the Stronghold when making your first portal.  
So I added a new column for it.

```python
df["One Portal Run"] = df["Second Portal"].isnull()
```

On low-food runs or before entering the End dimension, you’ll do what’s called a “hunger reset,” where you place a bed next to you to save your spawn
point and then kill yourself to fully restore your health and hunger bars, allowing you to continue the run.

```python
#Death Reset
df["intentional deaths"] = df["Total Deaths"] - df["Real_Deaths"]
```

When killing Blazes, you have a 50% chance of obtaining a Blaze Rod. Each rod can be crafted into Blaze Powder to make an Eye of Ender.
however, sometimes you can get a sword with Looting 1/2/3, which increases the chance of multiple rods dropping from a single Blaze.
I’ve created this column to identify which runs had a Looting sword.

```python
# Drop Rates
df["Blaze drop rate"] = df["Blaze Rods"] / df["Blaze Killed"]
df["Looting Run"] = df["Blaze drop rate"] > 1
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["IGT_minutes"],
    bins=30,
    kde=True,
    color="skyblue"
)

plt.title("Distribution of Run Times")
plt.xlabel("IGT (minutes)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### 1. Distribution of Run Times (Histogram)
The graph shows the distribution of run times (IGT) in minutes. It can be seen that most runs are concentrated in the range of approximately 18 to 25 minutes, with a peak around 21–22 minutes.

Additionally, there is a noticeable right tail, meaning there are a few very long runs (30+ minutes and beyond), indicating outliers where the run did not go as planned.

Conclusion:
Most runs are relatively consistent, but mistakes or bad RNG can lead to significantly higher outlier times.

```python
plt.figure(figsize=(10,5))

order = df.groupby("End Fight Type")["End Fight Minutes"].mean().sort_values().index

sns.boxplot(
    x="End Fight Type",
    y="End Fight Minutes",
    data=df,
    order=order,
    hue="End Fight Type",
    palette="pastel",
    legend=False
)

plt.title("End Fight Strategy vs Duration")
plt.xlabel("Strategy")
plt.ylabel("Time (minutes)")

plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

# Final Discussion & Conclusions:
After analyzing Couriway’s speedrun dataset, several clear patterns can be seen in what actually affects run performance.

The most important factor by far is player mistakes (Real Deaths). Runs with no deaths are consistently faster, while even a small number of deaths causes a noticeable increase in total run time. This shows that consistency and clean execution matter more than almost anything else in the run.

Another very important factor is the End fight strategy. Strategies like Zero Cycle and Ground Zero are significantly faster than older methods like Half Bow or Full Bow. This proves that optimized mechanics in the final fight have a huge impact on overall performance.

When comparing One Portal vs Two Portal routes, the data shows that using two portals is usually slightly faster on average.  in practice two portal routing allows better positioning and more efficient movement.

The analysis of Looting runs shows that Looting helps increase Blaze Rod drops and improves resource efficiency. However, it does not significantly reduce total run time, meaning that combat efficiency is not the main limiting factor in overall speed.

For Iron Source types, there are some differences in early-game Nether entry times, but they are relatively small compared to later stages of the run. This suggests that early RNG has less impact on final performance than execution and strategy later in the run.

Finally, when looking at performance over time, there is no strong improvement trend. Instead, run times fluctuate depending on mistakes, RNG, and individual run circumstances.

# Overall Conclusion

In general, the biggest factors affecting speedrun performance are execution quality and decision-making during the run, especially avoiding deaths and using optimal End fight strategies. RNG-based elements like iron source or loot drops play a smaller role in the final outcome compared to player skill and consistency.
